```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def x1(u):
    return -np.log(u) * 2


fig, ax = plt.subplots()
valores = np.random.random(100000)
val = x1(valores)
print(np.average(val))
ax.hist(val, bins = 100)
plt.show()
```

```python
def x2(tam):
    return np.random.gamma(shape=3, scale=1/0.4, size=tam)

fig, ax = plt.subplots()
val = x2(100000)
print(np.average(val))
ax.hist(val, bins = 100)
plt.show()
```

```python
def x3(u):
    return np.sin(np.pi * u)

fig, ax = plt.subplots()
valores = np.random.uniform(-1, 1, 100000)
val = x3(valores)
print(np.average(val))
ax.hist(val, bins = 100)
plt.show()
```

```python
def x4(tam):
    return np.random.lognormal(mean=0, sigma=1, size=tam)

fig, ax = plt.subplots()
val = x4(100000)
print(np.average(val))
ax.hist(val, bins = 100)
plt.show()
```

```python
def tablaMC():
    yp = []
    ip = []
    y = []
    Y = []
    for i in range(10):
        valx1 = x1(np.random.random(100000))
        valx2 = x2(100000)
        valx3 = x3(np.random.uniform(-1, 1, 100000))
        valx4 = x4(100000)
        for j in range(100000):
            y += [valx1[j]+valx2[j]+valx3[j]+valx4[j]]
        print(np.average(np.array(y)))
        yp += [np.average(np.array(y))]
        print(np.std(np.array(y), ddof=1))
        ip += [np.std(np.array(y), ddof=1)]
        Y += y
        y=[]
    print(np.average(np.array(yp)))
    print(np.average(np.array(ip)))
    fig, ax = plt.subplots()
    ax.hist(Y, bins=100)
    plt.show()

tablaMC()
```
