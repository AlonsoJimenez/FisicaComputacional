```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def obtenerDatos():
    datos = pd.read_excel('DatosEstrellas.xlsx', header=0, nrows=6220)
    return datos

datosEstelares = obtenerDatos()
```

```python
def sacarBV():
    bv = datosEstelares['BV']
    return bv

def sacarP():
    p = datosEstelares['p']
    return p

def sacar_m():
    m = datosEstelares['V']
    return m

BV = np.array(sacarBV())
p = np.array(sacarP())
m = np.array(sacar_m())

def calcularM():
    M = m + 5 * (np.log10(p) + 1)
    return M

M = calcularM()
```

```python


def graficarHR():
    fig, ax = plt.subplots()
    ax.scatter(BV, M, marker=".", c=np.arange(0, 100, 100/6220))
    ax.set_xlim(-0.45, 2.4)
    ax.set_ylabel('Magnitud absoluta')
    ax.set_xlabel('Clasificacin de color BV')
    ax.set_ylim(20, -10)
    fig.set_size_inches(18.5, 10.5)
    plt.show()

graficarHR()
```
